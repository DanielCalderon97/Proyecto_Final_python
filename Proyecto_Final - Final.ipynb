### 6. Crear una funcion para entrenar un modelo de regresion lineal de una variable y = β0 +β1 ∗x.

```python
# 6.1 Vector con la variable independiente x,

valorindependientex  = entrenamiento['OverallQual'].values.reshape(-1, 1) 


# 6.2 Vector con la variable dependiente y
valordependientey = entrenamiento['SalePrice'].values.reshape(-1, 1) 

# 6.3 Considerar el epochs como 1000

epochs = 1000

# 6.4 Definir un entero imprimr_error_cada

imprimir_error_cada = 10

# 6.5 Definir un entero learning_rate

learning_rate = 0.01   # Tasa de aprendizaje de 0.01 para que el descenso de gradiente sea optimo y no se quede atascado en un mínimo local




```

### 7. Consideraciones de modelo 

```python
# 7.1 Crear una matriz donde X sea una columna de unos y segunda columna de la matriz para todas las filas es igual a 1.


matrizX = np.ones((valorindependientex.shape[0], 2))  # Crear una matriz de unos con el mismo número de filas que valorindependientex y 2 columnas
matrizX[:, 0] = valorindependientex.flatten()  # Primera columna = datos de x
matrizX[:, 1] = 1  # Segunda columna = unos 


print(matrizX)
```

```python
#7.2 Inicializar los par´ametros del modelo en un vector β0 y β1 de ceros.

b0 = 0  # Intercepto
b1 = 0  # Pendiente
```

```python
# Definir el vector de parámetros (β1 y β0)
parametros = np.array([b1, b0])  # Nota: el orden es importante [pendiente, intercepto]

# Calcular las predicciones yˆ usando multiplicación matricial
y_pred = np.dot(matrizX, parametros)
```

```python

# 7.3 Mayores detalles


# Vector y verdadero
y_true = valordependientey.flatten()

# Crear vector para almacenar los errores (MSE) de cada iteración
errores_por_iteracion = np.zeros(epochs)

# Bucle de entrenamiento
for epoch in range(1, epochs + 1):
    # 1. Calcular predicciones (vectorizado)
    y_pred = np.dot(matrizX, [b1, b0])
    
    # 2. Calcular errores
    errores = y_pred - y_true
    
    # 3. Calcular el MSE (Mean Squared Error) para esta iteración
    mse = np.mean(errores**2)
    
    # 4. Almacenar el error en el vector
    errores_por_iteracion[epoch-1] = mse  # epoch-1 porque los índices empiezan en 0
    
    # 5. Calcular gradientes (derivadas parciales)
    gradiente_b1 = (1/len(y_true)) * np.dot(matrizX[:, 0], errores)
    gradiente_b0 = (1/len(y_true)) * np.sum(errores)
    
    # 6. Actualizar parámetros
    b1 = b1 - learning_rate * gradiente_b1
    b0 = b0 - learning_rate * gradiente_b0
    
    # 7. Mostrar error periódicamente
    if epoch % imprimir_error_cada == 0:
        print(f"Iteración {epoch}: Error (MSE) = {mse:.2f}")

# Mostrar parámetros finales
print("\nParámetros finales del modelo:")
print(f"β1 (Pendiente): {b1:.4f}")
print(f"β0 (Intercepto): {b0:.4f}")

# El vector errores_por_iteracion ahora contiene todos los errores
print("\nEjemplo de los primeros 10 errores:", errores_por_iteracion[:10])
```

```python
# Actualizar los parametros del modelo con la expresion:




# Preparación de datos
X = matrizX  # Matriz con columna de x y columna de unos
y = y_true    # Valores reales

# Vector para almacenar errores
errores_por_iteracion = np.zeros(epochs)

# Bucle de entrenamiento
for epoch in range(epochs):
    # 1. Calcular predicciones (vectorizado)
    y_pred = np.dot(X, parametros)  # X * [β1, β0] = β1*x + β0*1
    
    # 2. Calcular errores
    errores = y_pred - y
    
    # 3. Calcular gradientes (derivadas parciales)
    gradientes = np.array([
        (1/len(y)) * np.dot(X[:, 0], errores),  # δe/δβ1
        (1/len(y)) * np.sum(errores)            # δe/δβ0
    ])
    
    # 4. ACTUALIZACIÓN VECTORIZADA DE PARÁMETROS (esto es lo que necesitas)
    parametros = parametros - learning_rate * gradientes
    
    # 5. Almacenar error
    errores_por_iteracion[epoch] = np.mean(errores**2)
    
    # 6. Mostrar progreso (opcional)
    if (epoch+1) % 100 == 0:
        print(f"Iteración {epoch+1}: β1={parametros[0]:.4f}, β0={parametros[1]:.4f}, Error={errores_por_iteracion[epoch]:.2f}")

# Extraer los parámetros finales
b1, b0 = parametros
```

```python
# Entrenamiento del modelo de regresión lineal univariable con descenso de gradiente incluyendo la grafica de error en cada iteración

def entrenar_modelo(x, y, epochs=1000, imprimir_error_cada=100, alpha=0.01):
    """
    Entrena un modelo de regresión lineal univariable
    
    Args:
        x (np.array): Variable independiente
        y (np.array): Variable dependiente
        epochs (int): Número de iteraciones
        imprimir_error_cada (int): Frecuencia para imprimir progreso
        alpha (float): Tasa de aprendizaje
        
    Returns:
        tuple: (historial_modelo, errores_por_iteracion)
            - historial_modelo: dict {iteración: [β1, β0]}
            - errores_por_iteracion: np.array con errores en cada iteración
    """
    # Inicialización
    parametros = np.array([0.0, 0.0])  # [β1, β0]
    matrizX = np.column_stack((x, np.ones_like(x)))
    errores_por_iteracion = np.zeros(epochs)
    historial_modelo = {}
    
    for epoch in range(1, epochs + 1):
        # Calcular predicciones
        y_pred = np.dot(matrizX, parametros)
        
        # Calcular error y almacenar
        errores = y_pred - y
        mse = np.mean(errores**2)
        errores_por_iteracion[epoch-1] = mse
        
        # Calcular gradientes
        gradientes = np.array([
            (1/len(y)) * np.dot(matrizX[:, 0], errores),  # δe/δβ1
            (1/len(y)) * np.sum(errores)                  # δe/δβ0
        ])
        
        # Actualizar parámetros
        parametros = parametros - alpha * gradientes
        
        # Almacenar estado del modelo
        historial_modelo[epoch] = parametros.copy()
        
        # Imprimir progreso
        if epoch % imprimir_error_cada == 0 or epoch == 1:
            print(f"Iteración {epoch}: Error={mse:.2f}, β1={parametros[0]:.4f}, β0={parametros[1]:.4f}")
    
    return historial_modelo, errores_por_iteracion
```

```python
def visualizar_evolucion_error(errores_por_iteracion, titulo="Evolución del Error durante el Entrenamiento"):
    """
    Visualiza cómo cambia el error durante el entrenamiento
    
    Args:
        errores_por_iteracion (np.array): Vector de errores por iteración
        titulo (str): Título opcional para el gráfico
    """
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(errores_por_iteracion)+1), errores_por_iteracion, 'b-', linewidth=1)
    
    plt.xlabel('Número de Iteración', fontsize=12)
    plt.ylabel('Error Cuadrático Medio (MSE)', fontsize=12)
    plt.title(titulo, fontsize=14)
    
    # Mejorar visualización
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xlim(1, len(errores_por_iteracion))
    
    # Resaltar puntos clave
    min_error_idx = np.argmin(errores_por_iteracion)
    plt.scatter(min_error_idx+1, errores_por_iteracion[min_error_idx], 
                color='red', label=f'Mínimo error: {errores_por_iteracion[min_error_idx]:.2f}')
    
    plt.legend()
    plt.tight_layout()
    plt.show()
```

# 11. Crear una funci´on que nos permita visualizar con matplotlib c´omo evoluciona el modelo entrenado en el tiempo

```python
def visualizar_evolucion_modelo(historial_modelo, x_entrenamiento, y_entrenamiento, n=3):
    """
    Visualiza la evolución del modelo durante el entrenamiento
    
    Args:
        historial_modelo (dict): Diccionario con el historial de parámetros {iteración: [β1, β0]}
        x_entrenamiento (np.array): Datos de entrenamiento (variable independiente)
        y_entrenamiento (np.array): Datos de entrenamiento (variable dependiente)
        n (int): Cada cuántas iteraciones graficar el modelo
    """
    plt.figure(figsize=(12, 8))
    
    # Graficar datos de entrenamiento
    plt.scatter(x_entrenamiento, y_entrenamiento, color='blue', alpha=0.5, label='Datos reales')
    
    # Determinar iteraciones a graficar
    iteraciones = sorted(historial_modelo.keys())
    iteraciones_a_graficar = [it for it in iteraciones if it % n == 0 or it == iteraciones[-1]]
    
    # Paleta de colores para las rectas
    colors = plt.cm.rainbow(np.linspace(0, 1, len(iteraciones_a_graficar)))
    
    # Graficar cada modelo seleccionado
    for i, iteracion in enumerate(iteraciones_a_graficar):
        β1, β0 = historial_modelo[iteracion]
        x_min, x_max = np.min(x_entrenamiento), np.max(x_entrenamiento)
        x_line = np.array([x_min, x_max])
        y_line = β1 * x_line + β0
        
        plt.plot(x_line, y_line, color=colors[i], 
                 linestyle='-', linewidth=1.5, 
                 label=f'Iter. {iteracion}: y = {β1:.2f}x + {β0:.2f}')
    
    # Configuración del gráfico
    plt.xlabel('Variable Independiente (x)', fontsize=12)
    plt.ylabel('Variable Dependiente (y)', fontsize=12)
    plt.title(f'Evolución del Modelo (Mostrando cada {n} iteraciones)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.3)
    
    # Leyenda fuera del gráfico
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# Ejemplo de uso:
# Asumiendo que ya tienes:
# - historial_modelo (resultado de la función de entrenamiento)
# - x_entrenamiento (tus datos x de entrenamiento)
# - y_entrenamiento (tus datos y de entrenamiento)

# Visualizar evolución cada 100 iteraciones
visualizar_evolucion_modelo(
    historial_modelo=historial,
    x_entrenamiento=entrenamiento['OverallQual'].values,
    y_entrenamiento=entrenamiento['SalePrice'].values,
    n=100
)
```

```python
def entrenar_y_evaluar_modelos(variables_x, variable_y, df_entrenamiento, epochs_max=5000, alpha_inicial=0.01):
    """
    Entrena y evalúa modelos de regresión lineal para cada variable independiente
    
    Args:
        variables_x (list): Lista de nombres de variables independientes a evaluar
        variable_y (str): Nombre de la variable dependiente
        df_entrenamiento (pd.DataFrame): DataFrame con los datos de entrenamiento
        epochs_max (int): Número máximo de iteraciones a probar
        alpha_inicial (float): Tasa de aprendizaje inicial
        
    Returns:
        dict: Diccionario con los mejores modelos para cada variable
    """
    mejores_modelos = {}
    
    for var_x in variables_x:
        print(f"\n{'='*50}")
        print(f"Entrenando modelo para {var_x} vs {variable_y}")
        print(f"{'='*50}")
        
        # Preparar datos
        x = df_entrenamiento[var_x].values
        y = df_entrenamiento[variable_y].values
        
        # Experimentar con diferentes α
        alphas = [alpha_inicial, alpha_inicial*10, alpha_inicial/10]
        mejor_error = float('inf')
        mejor_modelo = None
        mejor_alpha = None
        mejores_epochs = None
        
        for alpha in alphas:
            print(f"\nProbando α = {alpha:.5f}")
            
            # Entrenar modelo con más epochs de las necesarias
            historial, errores = entrenar_modelo(
                x=x,
                y=y,
                epochs=epochs_max,
                imprimir_error_cada=epochs_max//10,
                alpha=alpha
            )
            
            # Analizar curva de aprendizaje
            epochs_optimos = determinar_epochs_optimos(errores)
            error_final = errores[epochs_optimos-1]
            
            # Graficar curva de aprendizaje
            visualizar_evolucion_error(
                errores[:epochs_optimos],
                titulo=f"Curva aprendizaje {var_x} (α={alpha:.5f})"
            )
            
            # Graficar evolución del modelo
            visualizar_evolucion_modelo(
                {k: v for k, v in historial.items() if k <= epochs_optimos},
                x, y,
                n=max(1, epochs_optimos//5)
            )
            
            # Actualizar mejor modelo si corresponde
            if error_final < mejor_error:
                mejor_error = error_final
                mejor_modelo = {
                    'β1': historial[epochs_optimos][0],
                    'β0': historial[epochs_optimos][1],
                    'errores': errores[:epochs_optimos],
                    'historial': {k: v for k, v in historial.items() if k <= epochs_optimos}
                }
                mejor_alpha = alpha
                mejores_epochs = epochs_optimos
        
        # Almacenar mejor modelo para esta variable
        mejores_modelos[var_x] = mejor_modelo
        print(f"\nMejor modelo para {var_x}:")
        print(f"α = {mejor_alpha:.5f}, epochs = {mejores_epochs}")
        print(f"Ecuación: y = {mejor_modelo['β1']:.4f}x + {mejor_modelo['β0']:.4f}")
        print(f"Error final (MSE): {mejor_error:.2f}")
    
    return mejores_modelos

def determinar_epochs_optimos(errores, tolerancia=1e-5, ventana=10):
    """
    Determina el número óptimo de epochs basado en la convergencia del error
    
    Args:
        errores (np.array): Vector de errores por iteración
        tolerancia (float): Umbral para considerar convergencia
        ventana (int): Tamaño de ventana para analizar estabilización
        
    Returns:
        int: Número óptimo de iteraciones
    """
    # Buscar si el error comienza a aumentar (sobreajuste)
    for i in range(len(errores)-1):
        if errores[i+1] > errores[i] * 1.01:  # Si el error aumenta un 1%
            return i+1
    
    # Si no hay sobreajuste, buscar convergencia
    if len(errores) > ventana:
        ultimos_errores = errores[-ventana:]
        variacion = np.max(ultimos_errores) - np.min(ultimos_errores)
        if variacion < tolerancia:
            return len(errores) - ventana
    
    return len(errores)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

def entrenar_con_sklearn(variables_x, variable_y, df_entrenamiento, df_validacion):
    """
    Entrena modelos lineales con scikit-learn para variables seleccionadas
    
    Args:
        variables_x (list): Lista de variables independientes
        variable_y (str): Variable dependiente
        df_entrenamiento (pd.DataFrame): Datos de entrenamiento
        df_validacion (pd.DataFrame): Datos de validación
        
    Returns:
        dict: Resultados de los modelos sklearn
    """
    resultados = {}
    
    for var_x in variables_x:
        print(f"\n{'='*50}")
        print(f"Modelo sklearn para {var_x} vs {variable_y}")
        print(f"{'='*50}")
        
        # Preparar datos
        X_train = df_entrenamiento[[var_x]].values
        y_train = df_entrenamiento[variable_y].values
        X_val = df_validacion[[var_x]].values
        y_val = df_validacion[variable_y].values
        
        # Estandarización (opcional pero recomendado)
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_val_scaled = scaler.transform(X_val)
        
        # Crear y entrenar modelo
        model = LinearRegression()
        model.fit(X_train_scaled, y_train)
        
        # Predecir y calcular métricas
        y_pred_train = model.predict(X_train_scaled)
        y_pred_val = model.predict(X_val_scaled)
        
        mse_train = mean_squared_error(y_train, y_pred_train)
        mse_val = mean_squared_error(y_val, y_pred_val)
        
        # Almacenar resultados
        resultados[var_x] = {
            'modelo': model,
            'scaler': scaler,
            'β1': model.coef_[0],
            'β0': model.intercept_,
            'mse_train': mse_train,
            'mse_val': mse_val
        }
        
        # Mostrar resultados
        print(f"\nEcuación: y = {model.coef_[0]:.4f}*x + {model.intercept_:.4f}")
        print(f"MSE entrenamiento: {mse_train:.2f}")
        print(f"MSE validación: {mse_val:.2f}")
        
        # Visualizar resultados
        plt.figure(figsize=(12, 5))
        
        # Gráfico de dispersión con recta de regresión
        plt.subplot(1, 2, 1)
        plt.scatter(X_train, y_train, alpha=0.5, label='Datos entrenamiento')
        x_line = np.array([[X_train.min()], [X_train.max()]])
        x_line_scaled = scaler.transform(x_line)
        y_line = model.predict(x_line_scaled)
        plt.plot(x_line, y_line, 'r-', linewidth=2, label='Regresión sklearn')
        plt.xlabel(var_x)
        plt.ylabel(variable_y)
        plt.title(f'Regresión sklearn: {var_x} vs {variable_y}')
        plt.legend()
        
        # Comparación de errores
        plt.subplot(1, 2, 2)
        plt.bar(['Entrenamiento', 'Validación'], [mse_train, mse_val], color=['blue', 'orange'])
        plt.title('Comparación de MSE')
        plt.ylabel('Error Cuadrático Medio')
        
        plt.tight_layout()
        plt.show()
    
    return resultados
```

```python
# Variables seleccionadas (basadas en tu análisis previo)
variables_seleccionadas = ['OverallQual', '1stFlrSF']

# Entrenar modelos con sklearn
resultados_sklearn = entrenar_con_sklearn(
    variables_x=variables_seleccionadas,
    variable_y='SalePrice',
    df_entrenamiento=entrenamiento,
    df_validacion=validacion_pruebas
)
```

# 14. Para cada variable independiente x elegida, crear una funci´on que usando el modelo entrenado manualmente

```python
def comparar_y_combinar_predicciones(x_values, modelo_manual, modelo_sklearn, scaler=None):
    """
    Compara y combina predicciones de ambos modelos
    
    Args:
        x_values (np.array): Valores de x para predecir
        modelo_manual (dict): Modelo entrenado manualmente (debe contener 'β1' y 'β0')
        modelo_sklearn (LinearRegression): Modelo de sklearn
        scaler (StandardScaler): Scaler usado en sklearn (opcional)
        
    Returns:
        tuple: (pred_manual, pred_sklearn, pred_combinada)
    """
    # Convertir a array numpy si no lo es
    x_values = np.array(x_values).reshape(-1, 1)
    
    # 1. Predicción con modelo manual
    β1 = modelo_manual['β1']
    β0 = modelo_manual['β0']
    pred_manual = β1 * x_values.flatten() + β0
    
    # 2. Predicción con sklearn (con escalado si existe)
    if scaler:
        x_scaled = scaler.transform(x_values)
    else:
        x_scaled = x_values
        
    pred_sklearn = modelo_sklearn.predict(x_scaled)
    
    # 3. Predicción combinada (promedio)
    pred_combinada = (pred_manual + pred_sklearn) / 2
    
    return pred_manual, pred_sklearn, pred_combinada
```

# 15. Usando el 20 % de los datos de validacion 

```python
def evaluar_modelos_validacion(mejores_modelos, resultados_sklearn, df_validacion, variable_y):
    """
    Evalúa los modelos en el conjunto de validación y determina el mejor para cada variable
    
    Args:
        mejores_modelos (dict): Modelos entrenados manualmente
        resultados_sklearn (dict): Modelos de sklearn
        df_validacion (pd.DataFrame): Datos de validación
        variable_y (str): Nombre de la variable dependiente
        
    Returns:
        dict: Resultados de evaluación para cada variable
    """
    resultados_evaluacion = {}
    
    for var_x in mejores_modelos.keys():
        print(f"\n{'='*50}")
        print(f"Evaluando modelos para {var_x}")
        print(f"{'='*50}")
        
        # Preparar datos
        x_val = df_validacion[var_x].values.reshape(-1, 1)
        y_val = df_validacion[variable_y].values
        
        # Obtener modelos
        modelo_manual = mejores_modelos[var_x]
        modelo_sklearn = resultados_sklearn[var_x]['modelo']
        scaler = resultados_sklearn[var_x]['scaler']
        
        # Calcular predicciones
        pred_manual, pred_sklearn, pred_combinada = comparar_y_combinar_predicciones(
            x_val, modelo_manual, modelo_sklearn, scaler)
        
        # Calcular errores (MSE)
        mse_manual = mean_squared_error(y_val, pred_manual)
        mse_sklearn = mean_squared_error(y_val, pred_sklearn)
        mse_combinada = mean_squared_error(y_val, pred_combinada)
        
        # Almacenar resultados
        resultados_evaluacion[var_x] = {
            'mse_manual': mse_manual,
            'mse_sklearn': mse_sklearn,
            'mse_combinada': mse_combinada,
            'mejor_modelo': None,
            'mejor_mse': float('inf')
        }
        
        # Determinar el mejor modelo
        for modelo, mse in [('manual', mse_manual), 
                           ('sklearn', mse_sklearn), 
                           ('combinada', mse_combinada)]:
            if mse < resultados_evaluacion[var_x]['mejor_mse']:
                resultados_evaluacion[var_x]['mejor_modelo'] = modelo
                resultados_evaluacion[var_x]['mejor_mse'] = mse
        
        # Graficar errores
        plt.figure(figsize=(10, 6))
        modelos = ['Manual', 'scikit-learn', 'Combinado']
        mses = [mse_manual, mse_sklearn, mse_combinada]
        
        bars = plt.bar(modelos, mses, color=['blue', 'green', 'purple'])
        plt.xlabel('Modelo')
        plt.ylabel('Error Cuadrático Medio (MSE)')
        plt.title(f'Comparación de Errores para {var_x} en Validación')
        
        # Añadir valores encima de las barras
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.2f}',
                    ha='center', va='bottom')
        
        plt.grid(True, axis='y', linestyle='--', alpha=0.7)
        plt.show()
        
        # Mostrar conclusión
        print(f"\nConclusión para {var_x}:")
        print(f"El mejor modelo es: {resultados_evaluacion[var_x]['mejor_modelo']}")
        print(f"Con un MSE de: {resultados_evaluacion[var_x]['mejor_mse']:.2f}")
        print("-"*50)
    
    return resultados_evaluacion
```
